```python
ativo = 'ITUB3.SA'
acao = yf.download(ativo, start = '2018-07-26', end = '2019-07-26')
sma9 = acao['Adj Close'].rolling(9).mean()
sma21 = acao['Adj Close'].rolling(21).mean()
acao['MM9'] = sma9
acao['MM21'] = sma21
```

```python
acao[['Adj Close', 'MM9', 'MM21']].plot(figsize = (25,25))
```
